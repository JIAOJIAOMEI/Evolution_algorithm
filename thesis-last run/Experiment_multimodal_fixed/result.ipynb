```python
def show_graph(dataframe):
    sns.set_theme(style="whitegrid",font_scale=1.5)
    sns.set_style("ticks",{'axes.grid' : True})
    dist_cols = 2
    dist_rows = 4
    plt.figure(figsize=(7 * dist_cols, 7 * dist_rows))
    i = 1
    for func in [13, 14, 15, 16, 17, 18, 19]:
        data = dataframe[dataframe["fitness_function"] == func]
        plt.subplot(dist_rows, dist_cols, i)
        ax = sns.boxplot(data=data, y="fitness",x="fitness_function",hue="algorithm",hue_order=["Baseline","Baldwin","Lamarck"],width=0.2,palette="Set1")
        ax.set(xlabel=None)
        ax.set(xticklabels=[])
        ax.tick_params(bottom=False)  # remove the ticks
        sns.move_legend(ax, "upper left", bbox_to_anchor=(0.6,1),frameon=False)
        # plot the optimal value, bold and red
        plt.axhline(y=optimal[func-1], color='r', linestyle='-',linewidth=2)
        plt.title(f"function={func}")
        plt.xlabel("")
        i = i + 1
    plt.savefig(f"multimodal_fixed_rawdata.png", dpi=400,bbox_inches='tight')
    plt.show()

show_graph(dataframe=df)
```

```python
# the cretira is that the abs(fitness - optimal) < 0.0001 or fitness < optimal
def get_best_percent():
    # get the optimal value for each function
    optimal = [0, 0, 0, 0, 0, 0, 0, -418.98 * 100, 0, 0, 0, 0, 0, 1, 0.00030, -1.0316, 0.3979, 3, -3.86, -3.32, -10.1532,
           -10.4028, -10.5363]
    # get the best percentage for each function
    # make a dataframe to store the data, columns are function, algorithm, best percentage
    best_percent = pd.DataFrame(columns=["function", "algorithm", "best_percentage"])
    # for each function
    for func in range(14,24,1):
        # for each algorithm
        for alg in ["Baseline", "Baldwin", "Lamarck"]:
            # get the data for this function and algorithm
            data = df[(df["fitness_function"] == func) & (df["algorithm"] == alg)]
            # get the optimal value for this function
            opt = optimal[func - 1]
            # count the number of times the fitness is the best
            count = 0
            # for each row in the data
            for index, row in data.iterrows():
                # if the abs(fitness - optimal) < 0.0001 or fitness < optimal
                if abs(row["fitness"] - opt) < 0.0001 or row["fitness"] < opt:
                    # count + 1
                    count = count + 1
            # get the best percentage
            if len(data) == 0:
                best_percentage = 0
            else:
                best_percentage = count / len(data)
            # to percentage
            best_percentage = best_percentage * 100
            # add the data to the dataframe
            best_percent = best_percent.append({"function": func, "algorithm": alg, "best_percentage": best_percentage},
                                               ignore_index=True)
    return best_percent
percent = get_best_percent()
print(percent)
```

```python
# plot the best percentage
def show_graph_percent(dataframe):
    sns.set_theme(style="whitegrid",font_scale=1.5)
    sns.set_style("ticks",{'axes.grid' : True})
    sns.set_palette("Set1")
    plt.figure(figsize=(20, 10))
    ax = sns.barplot(data=dataframe, x="function", y="best_percentage", hue="algorithm",hue_order=["Baseline","Baldwin","Lamarck"],width=0.6)
    sns.move_legend(ax,bbox_to_anchor=(1,1),frameon=False,loc="upper left",borderaxespad=0,ncol=1)
    plt.xlabel("Function")
    plt.ylabel("Percentage(%)")
    plt.savefig(f"percentage_14_23.png", dpi=400,bbox_inches='tight')
show_graph_percent(dataframe=percent)
```

```python
# for each function, do max min normalization
def max_min_normalization(dataframe):
    # for each function
    for func in range(14,24, 1):
        # get the data for this function
        data = dataframe[dataframe["fitness_function"] == func]
        print(data.index)
        # get the max and min value
        max = data["fitness"].max()
        min = data["fitness"].min()
        # for each row in the data
        for index, row in data.iterrows():
            # do the normalization
            dataframe.loc[index, "fitness"] = (row["fitness"] - min) / (max - min)
    return dataframe
df_box = max_min_normalization(dataframe=df)
print(df_box)
```

```python
# plot the best percentage
def show_graph_box(dataframe):
    sns.set_theme(style="whitegrid",font_scale=1.5)
    sns.set_style("ticks",{'axes.grid' : True})
    sns.set_palette("Set1")
    plt.figure(figsize=(25, 10))
    ax = sns.boxplot(data=dataframe, x="fitness_function", y="fitness", hue="algorithm",hue_order=["Baseline","Baldwin","Lamarck"])
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1,1),frameon=False)
    plt.xlabel("Function")
    plt.ylabel("Fitness")
    plt.savefig(f"rawdata_14_23.png", dpi=400,bbox_inches='tight')
show_graph_box(dataframe=df_box)
```
