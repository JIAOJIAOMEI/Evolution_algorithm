```python
one = pd.read_csv("/Users/meijiaojiao/Desktop/Evolution_algorithm/Experiment17/Experiment1/solutions_dataframe1.csv",header=0,index_col=0)
two = pd.read_csv("/Users/meijiaojiao/Desktop/Evolution_algorithm/Experiment17/Experiment1/solutions_dataframe2.csv",header=0,index_col=0)
three = pd.read_csv("/Users/meijiaojiao/Desktop/Evolution_algorithm/Experiment17/Experiment1/solutions_dataframe3.csv",header=0,index_col=0)
solution = pd.concat([one,two,three],axis=0)
print(colorama.Fore.YELLOW+f"the shape of solution is {solution.shape}")
solution.index = [i for i in range(0,solution.shape[0],1)]
solution.columns = ["Run"+str(i) for i in range(1,11,1)]
print(colorama.Fore.CYAN+f"the shape of solution dataframe is {solution.shape}")
print(colorama.Fore.CYAN+f"the columns of solution dataframe is the runs {solution.columns.tolist()}")
solution.to_csv("sd.csv",header=True,index=True)
print(solution.head(5))
solution = solution.stack()
solution = solution.rename_axis(["pm_index","Run"])
solution = solution.reset_index(name="fitness")
print(solution)
```

```python
# combine the two dataframes
# pm is the parameter combinations dataframe
# solution is the solutions dataframe
# according to pm_index to combine the two dataframes
pm["pm_index"] = [i for i in range(0,pm.shape[0],1)]
print(pm["pm_index"])
print(solution["pm_index"])
df = pd.merge(pm,solution,on="pm_index")
print(colorama.Fore.RED+f"the shape of df is {df.shape}")
print(df)
```

```python
def show_graph(dataframe):

    sns.set_theme(style="whitegrid",font_scale=1.5)
    sns.set_style("ticks",{'axes.grid' : True})
    dist_cols = 3
    dist_rows = 2
    plt.figure(figsize=(10 * dist_cols, 7 * dist_rows))
    i = 1
    for func in [14,15,16,17,18,19]:
        data = dataframe[dataframe["fitness_function"] == func]
        # normalize max and min
        data["fitness"] = (data["fitness"] - data["fitness"].min()) / (data["fitness"].max() - data["fitness"].min())
        plt.subplot(dist_rows, dist_cols, i)
        ax = sns.boxenplot(data=data, y="fitness",x="fitness_function",hue="algorithm",hue_order=["Baseline","Baldwin","Lamarck"],width=0.2,palette="Set2")
        sns.move_legend(ax, "upper left", bbox_to_anchor=(0.6,1),frameon=False)
        # 设置y的范围，突出显示数据最集中的地方
        # 先判断数据在哪里最集中，已经做了max min归一化的
        # 如果数据在0.5左右，那么就设置y的范围为0.4-0.6
        ax.set_ylim(data["fitness"].mean() - 0.03, data["fitness"].mean() + 0.03)
        plt.title(f"function={func}")
        i = i + 1
    plt.savefig(f"F141516171819_exp1.png", dpi=400,bbox_inches='tight')
    plt.show()

show_graph(dataframe=df)
```
