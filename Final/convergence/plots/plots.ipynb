```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
SSGA = pd.read_csv("/Users/meijiaojiao/Downloads/Final/convergence/SSGA/samples.txt",sep=",",header=None,index_col=[0])
index = SSGA.index
index = index.rename("Function")
SSGA.index = index
iterables =[[ i+1 for i in range(20)],["F"+str(i) for i in range(1,24,1)]]
index = pd.MultiIndex.from_product(iterables, names=["Combination","Function"])
SSGA.index = index
SSGA = SSGA.swaplevel(0,1)
SSGA
```

```python
Baldwin = pd.read_csv("/Users/meijiaojiao/Downloads/Final/convergence/Baldwin/samples.txt",sep=",",header=None,index_col=[0])
index = Baldwin.index
index = index.rename("Function")
Baldwin.index = index
iterables =[[ i+1 for i in range(20)],["F"+str(i) for i in range(1,24,1)]]
index = pd.MultiIndex.from_product(iterables, names=["Combination","Function"])
Baldwin.index = index
Baldwin = Baldwin.swaplevel(0,1)
Baldwin
```

```python
Lamarck = pd.read_csv("/Users/meijiaojiao/Downloads/Final/convergence/Lamarck/samples.txt",sep=",",header=None,index_col=[0])
index = Lamarck.index
index = index.rename("Function")
Lamarck.index = index
iterables =[[ i+1 for i in range(20)],["F"+str(i) for i in range(1,24,1)]]
index = pd.MultiIndex.from_product(iterables, names=["Combination","Function"])
Lamarck.index = index
Lamarck = Lamarck.swaplevel(0,1)
Lamarck
```

```python
func_list =["F"+str(i) for i in range(1,24,1)]
df = pd.concat([SSGA,Lamarck,Baldwin],keys=["SSGA","Lamarck","Baldwin"])
df = df.stack()
df = df.rename_axis(["Mode","Function","Combination","Iter"])
df = df.reset_index(level=[0,1],name="Best solution")
df["Best solution"] = pd.to_numeric(df["Best solution"])
df = df.reset_index(drop=False)
df
```

```python
# func_list =["F21","F22","F23","F1"]
def show_graph(dataframe):
    sns.set_theme(style="whitegrid",font_scale=1.5)
    sns.set_style("ticks",{'axes.grid' : True})
    dist_cols = 3
    dist_rows = 8
    plt.figure(figsize=(8 * dist_cols, 10 * dist_rows))
    i = 1
    for func in func_list:
        palette = {"SSGA":"lime",
           "Lamarck":"magenta",
           "Baldwin":"blue"}
        sns.set_theme(style="whitegrid",font_scale=2)
        data = dataframe[dataframe["Function"] == func]
        plt.subplot(dist_rows, dist_cols, i)
        ax =sns.lineplot(data=data,y="Best solution",x="Iter",hue="Mode",ci=95,err_style="band",linewidth=3.5,palette=palette,alpha=0.7)
        sns.move_legend(ax, "upper left", bbox_to_anchor=(0.6, 1),frameon=False)
        # plt.legend([],[], frameon=False)
        plt.xlabel("Iterations/50")
        plt.ylabel(f"Best solution")
        plt.title(f"{func}",fontsize=20, fontweight='bold')
        i = i + 1
    plt.savefig("./convergence.pdf", dpi=1200,bbox_inches='tight')
    plt.show()
show_graph(dataframe=df)
```
