```python
from matplotlib import cm
from matplotlib.colors import LightSource
from matplotlib.ticker import LinearLocator
def plot_xy(f):
    func, num_genes, genotype_range = functions_parameter.choose_func(f)
    z = []
    x,y = xy(f)
    f = "F"+str(f)
    for xi in x:
        for yi in y:
            if compare_string(f,"F15")==0 and compare_string(f,"F19")==0 and compare_string(f,"F20")==0 and compare_string(f,"F21")==0 and compare_string(f,"F22")==0 and compare_string(f,"F23")==0:
                z.append([func([xi,yi])])
            elif compare_string(f,"F19")==1:
                z.append([func([xi,yi,0])])
            elif compare_string(f,"F20")==1:
                z.append([func([xi,yi,0,0,0,0])])
            elif compare_string(f,"F15")==1 or compare_string(f,"F21")==1 or compare_string(f,"F22")==1 or compare_string(f,"F23")==1:
                z.append([func([xi,yi,0,0])])
    offset_value = min(z)
    print(offset_value[0])
    X, Y = np.meshgrid(x, y)
    shape = np.array(X).shape
    Z = np.array(z).reshape(shape)
    # plt.rcParams["font.family"] = "Times New Roman"
    # fig = plt.figure(figsize = (10,8))
    # ax = plt.axes(projection='3d')
    # surf = ax.plot_surface(X,Y,Z, cmap="viridis", lw=1,rstride=1, cstride=1,linewidth=0.5) #plt.cm.YlGnBu_r
    # ax.set_xlabel('X1', fontsize=14, fontweight='bold')
    # ax.set_ylabel('X2',fontsize=14, fontweight='bold')
    # ax.set_zlabel(f+"(X1,X2)",fontsize=14, fontweight='bold')
    # ax.set_title(f+": two dimensional view",fontsize=16, fontweight='bold')
    # ax.contour(X, Y, Z, zdir='z',offset=offset_value[0],cmap="viridis",linestyles="solid")
    # fig.colorbar(surf)
    # plt.savefig(f"{f}.png", dpi=1200 ,bbox_inches='tight')
    # plt.show()
    return X,Y,Z,offset_value
# for i in range(1,24,1):
#     plot_xy(i)
function_list = [2,4,5,7,8,9,12,13,14,16,19,22]
def show_graph(function_list):
    dist_cols = 4
    dist_rows = 3
    plt.figure(figsize=(dist_cols*5,dist_rows*5))
    i = 1
    for k in function_list:
        plt.rcParams["font.family"] = "Times New Roman"
        ax = plt.subplot(dist_rows, dist_cols, i,projection='3d')
        X,Y,Z,offset_value = plot_xy(k)
        ax.plot_surface(X,Y,Z, cmap="viridis") #plt.cm.YlGnBu_r ,rstride=1, cstride=1,linewidth=0.5
        ax.set_xlabel('X1', fontsize=12, fontweight='bold')
        ax.set_ylabel('X2',fontsize=12, fontweight='bold')
        ax.set_zlabel("F"+str(k)+"(X1,X2)",fontsize=12, fontweight='bold')
        ax.set_title("F"+str(k)+": two dimensional view",fontsize=12, fontweight='bold')
        ax.contour(X, Y, Z, zdir='z',offset=offset_value[0],cmap="viridis",linestyles="solid")
        i = i + 1
    plt.savefig("./Part of Functions plot.png",dpi=1200 ,bbox_inches='tight')
    plt.show()
show_graph(function_list)
```
