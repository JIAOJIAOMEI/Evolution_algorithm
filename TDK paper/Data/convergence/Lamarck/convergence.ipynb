```python
import pandas as pd
import numpy as np
```

```python
dfl = pd.read_csv("/Users/meijiaojiao/Downloads/Final/Lamarck/samples.txt",sep=",",header=None)
runs = dfl.shape[0]
index_ = ["RUN"+str(i+1) for i in range(runs)]
dfl.index = index_
dfl = dfl.T
dfl=dfl.drop(dfl.index[-1])
dfl = dfl.T
dfl
```

```python
dfb = pd.read_csv("/Users/meijiaojiao/Downloads/Final/Baldwin/samples.txt",sep=",",header=None)
index_ = ["RUN"+str(i+1) for i in range(runs)]
dfb.index = index_
dfb = dfb.T
dfb=dfb.drop(dfb.index[-1])
dfb = dfb.T
dfb
```

```python
dfs = pd.read_csv("/Users/meijiaojiao/Downloads/Final/SSGA/samples.txt",sep=",",header=None)
index_ = ["RUN"+str(i+1) for i in range(runs)]
dfs.index = index_
dfs = dfs.T
dfs=dfs.drop(dfs.index[-1])
dfs = dfs.T
dfs
```

```python
df = pd.concat([dfl,dfb,dfs],keys=["Lamarck","Baldwin","SSGA"])
df=df.stack()
df = df.rename_axis(["Mode","RUN","iter"])
df = df.reset_index(level=[0,1],name="best solution")
df["best solution"] = pd.to_numeric(df["best solution"])
df = df.reset_index(drop=False)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
mode_list=["Lamarck","Baldwin","SSGA"]
def show_graph(dataframe):
    sns.set_theme(style="whitegrid",font_scale=1.5)
    sns.set_style("ticks",{'axes.grid' : True})
    dist_cols = 1
    dist_rows = 1
    plt.figure(figsize=(15 * dist_cols, 5 * dist_rows))
    i = 1
    # for mode in mode_list:
    # dataf = dataframe[dataframe["Mode"] == mode]
    plt.subplot(dist_rows, dist_cols, i)
    ax =sns.lineplot(data=dataframe,y="best solution",x="iter",hue="Mode",ci=95)
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1),frameon=False)
    # plt.legend([],[], frameon=False)
    plt.xlabel("Iterations/100")
    plt.ylabel("best solution")
    # i = i + 1
    plt.show()
show_graph(dataframe=df)
```
